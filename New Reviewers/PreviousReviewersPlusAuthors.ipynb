```python
import pandas as pd
import numpy as np
import codecs
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

We obtain a file from the processing of the ACM Proceedings pages of LAK.  This code is in JAVA (ACMAuthorYear.java). 

```python
authorInfo=pd.read_csv('AuthorsInfo.csv',encoding="utf8")
authorInfo
```

Now we get the committee.csv file from Easychair to check who was in the previous PC

```python
previousPC=pd.read_csv('committee.csv',encoding="utf8")
previousPC['fullname']=previousPC["first name"]+" "+previousPC["last name"]
previousPC
```

```python
invitationPC=previousPC[['fullname','email','country','organization']]
invitationPC=invitationPC[previousPC['track #'] == 1]
#invitationPC=previousPC.loc[previousPC['track #'] == 1]
invitationPC["papers"]=0
invitationPC["years"]=""
toAdd=pd.DataFrame(columns=['fullname','email','country','organization','papers','years'])
```

```python

for index, row in authorInfo.iterrows():
    name=row["Author"]
    papers=row["Papers"]
    years=row["Years"]
    found=False
    for indexP, rowP in invitationPC.iterrows():
        namePrevious=rowP["fullname"]
        if fuzz.ratio(name, namePrevious)>80:
            invitationPC.loc[indexP,'papers']=papers
            invitationPC.loc[indexP,'years']=years
            found=True
            break
    if not found:
        toAdd.loc[len(toAdd)]=[name,"","","",papers,years]
        #invitationPC.append(df2)
        
    
    
```

```python
invitationPC=invitationPC.append(toAdd)
```
